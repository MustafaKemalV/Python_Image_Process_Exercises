```python
import numpy as np               
import matplotlib.pyplot as plt  

img = np.ones((600,600,3))

def DrawBezier(x0,y0,x1,y1,x2,y2,x3,y3):
    
    x = x0
    y = y0
    
    ts = 0.001
    
    t = 0
    while t <= 1.0:

        B0 = (1-t) * (1-t) * (1-t)
        B1 = 3 * (1-t) * (1-t) * t
        B2 = 3 * (1-t) * t * t
        B3 = t * t * t
        
        x = int(x0*B0 + x1*B1 + x2*B2 + x3*B3)
        y = int(y0*B0 + y1*B1 + y2*B2 + y3*B3)

        img[x,y] = (1,0,0)
        
        t=t+ts 
    
    return
   
DrawBezier(100, 100, 150, 350, 250, 350, 300, 100)
DrawBezier(300, 100, 390, 350, 410, 350, 500, 100)

    
%matplotlib notebook
plt.figure(figsize=(10,8), dpi = 100, facecolor='w')
plt.imshow(img)
plt.show()
```
