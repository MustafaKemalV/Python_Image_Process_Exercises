```python
import numpy as np
from numpy import asarray

import cv2
from PIL import Image

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_A = cv2.imread('C:\\Users\\musta\\Desktop\\Riga.jpg')
image_B = cv2.imread('C:\\Users\\musta\\Desktop\\Stockholm.jpg')

im_rgb_A = cv2.cvtColor(image_A, cv2.COLOR_BGR2RGB)
im_rgb_B = cv2.cvtColor(image_B, cv2.COLOR_BGR2RGB)

resized_img_A = cv2.resize(image_A, (500,500))
resized_img_B = cv2.resize(image_B, (500,500))

cv2.imwrite('C:\\Users\\musta\\Desktop\\Riga.jpg', resized_img_A)
cv2.imwrite('C:\\Users\\musta\\Desktop\\Stockholm.jpg', resized_img_B)

image_A = Image.open('C:\\Users\\musta\\Desktop\\Riga.jpg')
image_B = Image.open('C:\\Users\\musta\\Desktop\\Stockholm.jpg')

data_A = asarray(image_A)
data_B = asarray(image_B)

image_array_A = Image.fromarray(data_A)
image_array_B = Image.fromarray(data_B)

image_norm_A = data_A/255.0
image_norm_B = data_B/255.0

fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(image_norm_A)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(image_norm_B)

ColorBurn = 1 -  ((1 - image_norm_B) / image_norm_A)

fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(1,3,1)
ax1.title.set_text('input img 1')
ax1.imshow(image_norm_A)

ax2 = fig.add_subplot(1,3,2)
ax2.title.set_text('input img 2')
ax2.imshow(image_norm_B)

ax3 = fig.add_subplot(1,3,3)
ax3.title.set_text('ColorBurn')
ax3.imshow(ColorBurn)

HardLight = np.where(image_norm_A.all() <= 0.5, 2 * image_norm_A * image_norm_B, 1 - 2 * (1 - image_norm_A) * (1 - image_norm_B)) 

fig = plt.figure(figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(1,3,1)
ax1.title.set_text('input img 1')
ax1.imshow(image_norm_A)

ax2 = fig.add_subplot(1,3,2)
ax2.title.set_text('input img 2')
ax2.imshow(image_norm_B)

ax3 = fig.add_subplot(1,3,3)
ax3.title.set_text('HardLight')
ax3.imshow(HardLight)
```
