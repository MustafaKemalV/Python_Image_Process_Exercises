```python
import numpy as np
import matplotlib.pyplot as plt
import math

img = np.ones((1000,1000,3))

point1 = (300,100)
point2 = (100,100)
point3 = (100,300)
point4 = (300,300)

def MoveObject(move_x, move_y):
    
    x1 = point1[0] + move_x
    y1 = point1[1] + move_y
    x2 = point2[0] + move_x
    y2 = point2[1] + move_y
    x3 = point3[0] + move_x
    y3 = point3[1] + move_y
    x4 = point4[0] + move_x
    y4 = point4[1] + move_y
    
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)

def ScaleObject(scale_X, scale_Y):
    
    x1 = point1[0] * scale_X
    y1 = point1[1] * scale_Y
    x2 = point2[0] * scale_X
    y2 = point2[1] * scale_Y
    x3 = point3[0] * scale_X
    y3 = point3[1] * scale_Y
    x4 = point4[0] * scale_X
    y4 = point4[1] * scale_Y
    
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
def RotateObject(angle):
    
    x1 = point1[0] * math.cos(angle)-point1[1] * math.sin(angle)
    y1 = point1[0] * math.sin(angle)+point1[1] * math.cos(angle)
    x2 = point2[0] * math.cos(angle)-point2[1] * math.sin(angle)
    y2 = point2[0] * math.sin(angle)+point2[1] * math.cos(angle)
    x3 = point3[0] * math.cos(angle)-point3[1] * math.sin(angle)
    y3 = point3[0] * math.sin(angle)+point3[1] * math.cos(angle)
    x4 = point4[0] * math.cos(angle)-point4[1] * math.sin(angle)
    y4 = point4[0] * math.sin(angle)+point4[1] * math.cos(angle)
    
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    


plt.figure(figsize=(10,8), dpi = 100, facecolor='w')
plt.imshow(img)



MoveObject(120,100)
ScaleObject(3,3)
RotateObject(288744)


plt.show()
```
